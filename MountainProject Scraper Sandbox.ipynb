```python
#for state of Alamaba, print out first all areas within first stop_reg climbing regions
content = soup.find_all('div',class_='mb-half')
counter = 0
stop = 0
stop_reg = 1
for c in content:
    if counter > stop:
        break
    else:
        curr = c.find('a')
        state_link = curr['href']
        state_page = urlopen(state_link)
        state_soup = BeautifulSoup(state_page,'html.parser')
        
        #get climbing regions within the state
        regions_content = state_soup.find_all('div',class_='lef-nav-row')
        #print(regions_content)
        
        state_name = state_soup.find('h1').get_text()
        print(state_name)
        counter_reg = 0
        for rc in regions_content:
            if counter_reg > stop_reg:
                break
            else:
                curr_region = rc.find('a')
                curr_region_text = curr_region.get_text()
                curr_region_link = curr_region['href']
                print('-',curr_region_text)
                
                region_page = urlopen(curr_region_link)
                region_soup = BeautifulSoup(region_page,'html.parser')
                
                #get climbing areas within region
                area_content = region_soup.find_all('div',class_='lef-nav-row')
                if counter_reg == 1:
                    print(area_content)
                    
                area_name = region_soup.find('h1').get_text()
                print('--',area_name)
                
                for ac in area_content:
                    curr_area = ac.find('a')
                    curr_area_text = curr_area.get_text()
                    curr_area_link = curr_area['href']
                    print('---',curr_area_text,curr_area_link)
                    
                time.sleep(1)
                
            counter_reg += 1
            
    counter += 1
```

## Put all together

```python
from urllib.request import urlopen
from IPython.core.debugger import set_trace

#returns state/region/route area/route from tag with class name class_name

'''
input:
    soup: BeautifulSoup instance with html parser
    tag: html tag to find
    class_name: class name to find within tag
    id_name: id name of class
    stop_ind: routine returns first stop_ind 
    is_route_level: boolean, set to true if this level is the climbing routes
'''
def get_route_level_data(soup, tag, class_name, id_name='', stop_ind=-1, is_route_level=False):
    if not id_name:
        content = soup.find_all(tag,class_=class_name)
    else:
        content = soup.find_all(tag,class_=class_name,id=id_name)
    
    counter = 0
    
    link_list = []
    text_list = []
    for c in content:
        if is_route_level:
            #set_trace()
            routes = c.find_all('a')
            
            for r in routes:
                if stop_ind != -1:
                    if counter > stop_ind:
                        break
                
                curr_route_name = r.get_text()
                curr_route_link = r['href']
                
                link_list.append(curr_route_link)
                text_list.append(curr_route_name)
        else:
            if stop_ind != -1:
                if counter > stop_ind:
                    break
                    
            curr = c.find('a')
            curr_link = curr['href']
            curr_text = curr.get_text()
        
            link_list.append(curr_link)
            text_list.append(curr_text)
        
            counter += 1
        
    return content, link_list, text_list
```

```python
from bs4 import BeautifulSoup
from IPython.core.debugger import set_trace
import time

route_page = 'https://www.mountainproject.com/route-guide'
page = urlopen(route_page)
soup = BeautifulSoup(page,'html.parser')

#get data for Alabama only
content_route_dir, link_route_dir, text_route_dir = get_route_level_data(soup,'div','mb-half',stop_ind=0)

for rdind in range(len(link_route_dir)):
    print(text_route_dir[rdind], link_route_dir[rdind])
    
    time.sleep(1)
    state_soup = BeautifulSoup(urlopen(link_route_dir[rdind]),'html.parser')
    
    #get data for climbing regions within state
    content_region, link_region, text_region = get_route_level_data(state_soup,'div','lef-nav-row',stop_ind=2)
    
    for regind in range(len(link_region)):
        print('\t',text_region[regind],link_region[regind])
        
        time.sleep(1)
        reg_soup = BeautifulSoup(urlopen(link_region[regind]),'html.parser')
        
        content_area, link_area, text_area = get_route_level_data(reg_soup,'div','lef-nav-row')
        
        #if content area is empty, check for routes instead
        if not content_area:
            #set_trace()
            content_route, link_route, text_route = get_route_level_data(reg_soup,'table','width100',id_name='left-nav-route-table',is_route_level=True)
                
            #set_trace()
            for routeind in range(len(text_route)):
                print('\t\t\t',text_route[routeind],link_route[routeind])
        else:
            for areaind in range(len(content_area)):
                print('\t\t',text_area[areaind],link_area[areaind])
                
                time.sleep(1)
                area_soup = BeautifulSoup(urlopen(link_area[areaind]),'html.parser')
                
                #set_trace()
                content_route, link_route, text_route = get_route_level_data(area_soup,'table','width100',id_name='left-nav-route-table',is_route_level=True)
                
                for routeind in range(len(text_route)):
                    print('\t\t\t',text_route[routeind],link_route[routeind])
```

```python
def get_route_level_data_helper(soup, stop_ind=2):
    #set_trace()
    content_area, link_area, text_area = \
        get_route_level_data(soup,'div','lef-nav-row',stop_ind=stop_ind)
    
    content_route, link_route, text_route = \
        get_route_level_data(soup,'table','width100',id_name='left-nav-route-table',is_route_level=True,stop_ind=stop_ind)
    
    return content_area, link_area, text_area, content_route, link_route, text_route
```

```python
from bs4 import BeautifulSoup
from IPython.core.debugger import set_trace
import time
from urllib.request import urlopen

def get_children(soup):
    content_area, link_area, text_area, \
        content_route, link_route, text_route = get_route_level_data_helper(soup,stop_ind=-1)
    #set_trace()
    if link_area:
        #print(link_area)
        
        link_master = []
        text_master = []
    
        for laind in range(len(link_area)):            
            time.sleep(0.01)
            curr_soup = BeautifulSoup(urlopen(link_area[laind]),'html.parser')
            
            #set_trace()
            
            [link, text] = get_children(curr_soup)
            link_master.append([link_area[laind], link])
            text_master.append([text_area[laind], text])
            
        return [link_master, text_master]
    
    if link_route:
        return [link_route, text_route]
    else:
        return None
```
